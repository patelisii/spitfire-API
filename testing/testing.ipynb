```python
from mutagen.mp3 import MP3

def get_audio_length(audio_file):
    audio = MP3(audio_file)
    audio_length = audio.info.length
    return audio_length

audio_file = "trump_biden/verse0.mpeg"
length = get_audio_length(audio_file)
print(f"Audio length: {length} seconds")
```

```python
# LENGTH OF AN MP3 FILE

from pydub import AudioSegment

def get_audio_length(audio_file):
    audio = AudioSegment.from(audio_file)
    audio_length = len(audio) / 1000
    return audio_length

audio_file = "trump_biden/"
length = get_audio_length(audio_file)
print(f"Audio length: {length} seconds")
```

```python
def get_video_length(video_file):
    cap = cv2.VideoCapture(video_file)
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    video_length = frame_count / fps
    cap.release()
    return video_length

video_file = "donaldVid.mp4"
length = get_video_length(video_file)
print(f"Video length: {length} seconds")
```

```python
import cv2
import numpy as np

# Load the video files
video_file_1 = "donaldVid.mp4"
video_file_2 = "joeRoast.mp4"

def combine_vids(video_file_1, video_file_2):

    cap1 = cv2.VideoCapture(video_file_1)
    cap2 = cv2.VideoCapture(video_file_2)

    # Get video properties
    width1 = int(cap1.get(cv2.CAP_PROP_FRAME_WIDTH))
    height1 = int(cap1.get(cv2.CAP_PROP_FRAME_HEIGHT))
    width2 = int(cap2.get(cv2.CAP_PROP_FRAME_WIDTH))
    height2 = int(cap2.get(cv2.CAP_PROP_FRAME_HEIGHT))

    fps1 = int(cap1.get(cv2.CAP_PROP_FPS))
    fps2 = int(cap2.get(cv2.CAP_PROP_FPS))

    # Set the output video properties and create VideoWriter object
    output_width = width1 + width2
    output_height = max(height1, height2)
    output_fps = min(fps1, fps2)
    output_file = "outputVid.mp4"
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')

    out = cv2.VideoWriter(output_file, fourcc, output_fps, (output_width, output_height))

    while cap1.isOpened() and cap2.isOpened():
        ret1, frame1 = cap1.read()
        ret2, frame2 = cap2.read()

        if ret1 and ret2:
            # Resize frames to have the same height
            frame1_resized = cv2.resize(frame1, (width1, output_height))
            frame2_resized = cv2.resize(frame2, (width2, output_height))

            # Combine frames side by side
            combined_frame = np.hstack((frame1_resized, frame2_resized))

            # Write the combined frame to the output video
            out.write(combined_frame)

            # Display the combined frame (optional)
            cv2.imshow('Combined Video', combined_frame)

            if cv2.waitKey(1) & 0xFF== ord('q'):
                break
        else:
            break

    # Release resources
    cap1.release()
    cap2.release()
    out.release()
    cv2.destroyAllWindows()
```

```python
import ffmpeg

def add_background_music(video_file, music_file, output_file, volume=0.1):
    video = ffmpeg.input(video_file)
    music = ffmpeg.input(music_file)

    video_duration = float(ffmpeg.probe(video_file)['format']['duration'])
    music_duration = float(ffmpeg.probe(music_file)['format']['duration'])

    loop_count = int(video_duration // music_duration) + 1

    audio = ffmpeg.concat(*([music] * loop_count), v=0, a=1).filter("atrim", duration=video_duration)
    audio = audio.filter("volume", volume)

    video_stream = video.video
    audio_stream = video.audio

    mixed_audio = ffmpeg.filter([audio_stream, audio], "amix")

    output = ffmpeg.output(video_stream, mixed_audio, output_file, vcodec='copy', acodec="aac", strict="experimental")
    ffmpeg.run(output)

video_file = "path/to/your/video_file.mp4"
music_file = "path/to/your/background_music.mp3"
output_file = "path/to/your/output_video_file_with_music.mp4"

add_background_music("swift_tate.mp4", "beats/beat1.m4a", "test.m4a")
```

```python
from spitfire_video_editing import *


def mix_audio_files(main_audio_file, background_audio_file, output_file, background_volume=-10):
    main_audio = AudioSegment.from_mp3(main_audio_file)
    background_audio = AudioSegment.from_mp3(background_audio_file)

    # Set the background audio volume
    background_audio = background_audio + background_volume

    # Overlay the audio files
    combined_audio = main_audio.overlay(background_audio)

    # Export the output file
    combined_audio.export(output_file, format="mp3")


mix_audio_files("audio_temp.mp3", "beats/beat1.mp3", "test.mp3")
```

```python
def append_vids(video1, video2, output):
    # Create a temporary file with the list of input video files
    with open("input_list.txt", "w") as f:
        f.write(f"file '{video1}'\n")
        f.write(f"file '{video2}'\n")

    # Call ffmpeg to concatenate the video files
    command = f"ffmpeg -f concat -safe 0 -i input_list.txt -c copy {output}"
    os.system(command)

    if os.path.exists("input_list.txt"):
        os.remove("input_list.txt")
```
